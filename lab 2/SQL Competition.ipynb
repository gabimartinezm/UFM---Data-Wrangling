```python
#pregunta 1 - nombre del primer superheroe
query1 = " SELECT * FROM heroes_information limit 10"
pysqldf(query1)
```

```python
#pregunta 2 - numero de casas publicadoras
query2 = """
SELECT DISTINCT Publisher
FROM heroes_information;

"""

pysqldf(query2)
```

```python
#pregunta 3 - superheroes que miden más de 200 cms
query3 = """
SELECT COUNT(*)
FROM heroes_information
WHERE Height > 200;

"""

pysqldf(query3)
```

```python
#pregunta 4 - superheroes humanos y más de 200 cms
query4 = """
SELECT COUNT(*)
FROM heroes_information
WHERE Race='Human'
AND Height > 200;

"""

pysqldf(query4)
```

```python
#pregunta 5 - superheroes entre 100 lbs y 200 lbs
query5 = """
SELECT COUNT(*)
FROM heroes_information
WHERE Weight
BETWEEN 100 AND 200;

"""

pysqldf(query5)
```

```python
#pregunta 6 - superheroes con ojos azules o rojos 
query6 = """
SELECT COUNT(*)
FROM heroes_information
WHERE Eye_color ='blue' or Eye_color='red' 
"""

pysqldf(query6)
```

```python
#pregunta 7 - superheroes que son humanos o (mutant y tienen el pelo verde) o (vampiros con pelo negro)
query7 =  """
SELECT COUNT(*) 
FROM heroes_information
WHERE (Race = 'Human') OR 
      (Hair_color = 'Green' AND Race = 'Mutant')  OR (Race = 'Vampire' AND Hair_color = 'Black')
"""

pysqldf(query7)
```

```python
#pregunta 8 - primer superheore en orden descendente 
query8 =  """
SELECT *
FROM heroes_information order by name DESC;
"""

pysqldf(query8)
```

```python
#pregunta 9 - superheroes femeninos y masculinos
query9 =  """
SELECT gender, COUNT(*)
FROM heroes_information group by gender;
"""

pysqldf(query9)
```

```python
#pregunta 10 - casas publicadoras con más de 15 superheroes 
query10 =  """
SELECT COUNT(*)
FROM heroes_information 
GROUP BY Publisher
HAVING COUNT(Publisher) > 15;
"""

pysqldf(query10)
```
